# Python Exception Handling

## 1. Introduction to Exceptions

### What are Exceptions?
- Exceptions are events that disrupt the normal flow of a program's execution
- They occur when an error is encountered during program execution
- Python uses exceptions to handle errors gracefully
- When an exception occurs, Python stops the current execution and looks for an exception handler

### Why Exception Handling is Important?
- Prevents program crashes
- Provides meaningful error messages
- Allows graceful recovery from errors
- Improves code reliability and user experience

### Common Exception Types

| Exception Type | Description | Example |
|----------------|-------------|---------|
| `ValueError` | Raised when a function receives an argument of correct type but inappropriate value | `int("abc")` |
| `TypeError` | Raised when an operation is applied to an object of inappropriate type | `"2" + 2` |
| `NameError` | Raised when a local or global name is not found | `print(undefined_var)` |
| `IndexError` | Raised when sequence index is out of range | `list[10]` on list of 5 items |
| `KeyError` | Raised when dictionary key is not found | `dict["missing_key"]` |
| `ZeroDivisionError` | Raised when division by zero occurs | `10 / 0` |
| `FileNotFoundError` | Raised when file is not found | `open("nonexistent.txt")` |
| `AttributeError` | Raised when attribute reference fails | `"hello".nonexistent_method()` |

````python
# Examples of common exceptions

# ValueError
try:
    num = int("abc")
except ValueError as e:
    print(f"ValueError: {e}")

# TypeError
try:
    result = "2" + 2
except TypeError as e:
    print(f"TypeError: {e}")

# IndexError
try:
    my_list = [1, 2, 3]
    print(my_list[10])
except IndexError as e:
    print(f"IndexError: {e}")
````

````python
# More exception examples

# KeyError
try:
    my_dict = {"a": 1, "b": 2}
    print(my_dict["c"])
except KeyError as e:
    print(f"KeyError: {e}")

# ZeroDivisionError
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"ZeroDivisionError: {e}")

# AttributeError
try:
    text = "hello"
    print(text.nonexistent_method())
except AttributeError as e:
    print(f"AttributeError: {e}")
````

## 2. try-except Blocks

### Basic Syntax
```python
try:
    # Code that might raise an exception
except ExceptionType:
    # Code to handle the exception
```

### Handling Single Exceptions
- Specify the type of exception you want to catch
- Use `as` to capture the exception object
- The except block executes only if that specific exception occurs

````python
# Basic try-except example
def divide(a, b):
    try:
        result = a / b
        print(f"Result: {result}")
    except ZeroDivisionError:
        print("Error: Cannot divide by zero!")

print("Test 1: 10 / 2")
divide(10, 2)

print("\nTest 2: 10 / 0")
divide(10, 0)
````

### Handling Multiple Exceptions
- Use multiple except blocks for different exception types
- Python executes the first matching except block
- Order matters: more specific exceptions should come before general ones

````python
# Multiple except blocks
def process_input(value):
    try:
        num = int(value)
        result = 100 / num
        print(f"Result: {result}")
    except ValueError:
        print("Error: Please enter a valid integer")
    except ZeroDivisionError:
        print("Error: Cannot divide by zero")

print("Test 1: '10'")
process_input("10")

print("\nTest 2: 'abc'")
process_input("abc")

print("\nTest 3: '0'")
process_input("0")
````

### Catching Multiple Exceptions in One Block
- Use parentheses to catch multiple exception types in a single except block
- Useful when handling logic is the same for different exceptions

````python
# Catching multiple exceptions in one block
def safe_operation(data, index, key):
    try:
        value = data[index]
        result = value[key]
        print(f"Result: {result}")
    except (IndexError, KeyError) as e:
        print(f"Error: Invalid access - {e}")
    except TypeError as e:
        print(f"Error: Type mismatch - {e}")

print("Test 1: Valid access")
safe_operation([{"a": 1}, {"b": 2}], 0, "a")

print("\nTest 2: Invalid index")
safe_operation([{"a": 1}], 5, "a")

print("\nTest 3: Invalid key")
safe_operation([{"a": 1}], 0, "b")
````

### Catching Exception Objects
- Use `as` to assign the exception object to a variable
- Access error message and other exception attributes
- Useful for logging and debugging

````python
# Catching exception object
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError as e:
        print(f"Exception type: {type(e).__name__}")
        print(f"Exception message: {e}")
        print(f"Exception args: {e.args}")
        return None

result = safe_divide(10, 0)
print(f"Returned value: {result}")
````

## 3. else and finally Blocks

### else Block
- Executes if no exception occurs in the try block
- Must come after all except blocks
- Useful for code that should only run when try succeeds

### finally Block
- Always executes, regardless of whether an exception occurred
- Used for cleanup operations (closing files, releasing resources)
- Executes even if return statement is used in try/except

````python
# else block example
def divide_with_else(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("Cannot divide by zero")
    else:
        print(f"Division successful! Result: {result}")

print("Test 1: 10 / 2")
divide_with_else(10, 2)

print("\nTest 2: 10 / 0")
divide_with_else(10, 0)
````

````python
# finally block example
def process_file():
    file = None
    try:
        file = open("example.txt", "r")
        content = file.read()
        print(f"File content: {content}")
    except FileNotFoundError:
        print("File not found!")
    finally:
        if file:
            file.close()
            print("File closed in finally block")
        print("Finally block always executes")

process_file()
````

````python
# Complete try-except-else-finally structure
def complete_structure(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("Exception: Division by zero")
    else:
        print(f"No exception! Result: {result}")
    finally:
        print("Cleanup: Operation complete")

print("Test 1: 10 / 2")
complete_structure(10, 2)

print("\nTest 2: 10 / 0")
complete_structure(10, 0)
````

## 4. Raising Exceptions

### raise Keyword
- Used to explicitly raise an exception
- Can raise built-in exceptions or custom exceptions
- Useful for enforcing conditions or validating inputs

### Custom Error Messages
- Provide meaningful error messages when raising exceptions
- Helps with debugging and user understanding

````python
# Raising exceptions
def set_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative")
    if age > 150:
        raise ValueError("Age cannot be greater than 150")
    return age

try:
    print("Test 1: age = 25")
    print(set_age(25))
    
    print("\nTest 2: age = -5")
    print(set_age(-5))
except ValueError as e:
    print(f"Error: {e}")
````

### Re-raising Exceptions
- Use `raise` without arguments in an except block to re-raise the caught exception
- Useful when you want to perform some action but still let the exception propagate

````python
# Re-raising exceptions
def process_data(data):
    try:
        result = int(data)
        return result
    except ValueError:
        print(f"Warning: Could not convert '{data}' to integer")
        raise  # Re-raise the exception

try:
    process_data("abc")
except ValueError:
    print("Exception caught in outer block")
````

## 5. Custom Exceptions

### Creating Exception Classes
- Inherit from the `Exception` class or its subclasses
- Define custom exception classes for specific error conditions
- Makes code more readable and maintainable

### When to Use Custom Exceptions
- When you need to handle specific error conditions differently
- When built-in exceptions don't capture the error semantics
- When building libraries or frameworks

````python
# Custom exception class
class InsufficientFundsError(Exception):
    """Raised when account has insufficient funds"""
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount = amount
        super().__init__(f"Insufficient funds: balance={balance}, amount={amount}")

class BankAccount:
    def __init__(self, balance):
        self.balance = balance
    
    def withdraw(self, amount):
        if amount > self.balance:
            raise InsufficientFundsError(self.balance, amount)
        self.balance -= amount
        return self.balance

# Using custom exception
account = BankAccount(100)
try:
    print(f"Initial balance: {account.balance}")
    account.withdraw(50)
    print(f"After withdrawal: {account.balance}")
    account.withdraw(100)  # This will raise exception
except InsufficientFundsError as e:
    print(f"Error: {e}")
````

## 6. Exception Handling Best Practices

### Specific vs Generic Exceptions
- **DO**: Catch specific exceptions you expect
- **DON'T**: Catch generic `Exception` unless necessary
- Catching generic exceptions can hide bugs and make debugging difficult

````python
# Bad practice: Catching generic Exception
def bad_practice():
    try:
        result = 10 / 0
    except Exception:  # Too generic
        print("Something went wrong")  # Not helpful

# Good practice: Catching specific exception
def good_practice():
    try:
        result = 10 / 0
    except ZeroDivisionError:  # Specific
        print("Cannot divide by zero")  # Clear and helpful

print("Bad practice:")
bad_practice()

print("\nGood practice:")
good_practice()
````

### Logging Exceptions
- Use logging module to log exceptions
- Include traceback information for debugging
- Log at appropriate levels (ERROR, WARNING, INFO)

````python
import traceback

# Logging exception with traceback
def risky_operation():
    try:
        result = 10 / 0
    except ZeroDivisionError:
        print("Error occurred:")
        print(traceback.format_exc())  # Print full traceback

risky_operation()
````

### Context Managers (with statement)
- Use context managers for automatic resource cleanup
- Eliminates need for explicit finally blocks
- More Pythonic and safer

````python
# Using context manager for file handling
def read_file_safely(filename):
    try:
        with open(filename, "r") as file:
            content = file.read()
            print(f"File content: {content}")
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found")
    # File is automatically closed, even if exception occurs

# This will fail (file doesn't exist), but demonstrates the pattern
read_file_safely("nonexistent.txt")
````

## 7. Practical Examples

### File Operations
Handling file operations with proper exception handling.

````python
# File operation with exception handling
def read_and_process_file(filename):
    try:
        with open(filename, "r") as file:
            content = file.read()
            lines = content.splitlines()
            return len(lines)
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found")
        return 0
    except PermissionError:
        print(f"Error: No permission to read '{filename}'")
        return 0
    except Exception as e:
        print(f"Unexpected error: {e}")
        return 0

line_count = read_and_process_file("nonexistent.txt")
print(f"Lines read: {line_count}")
````

### User Input Validation
Validating user input with exception handling.

````python
# User input validation
def get_positive_integer(prompt):
    while True:
        try:
            value = int(input(prompt))
            if value < 0:
                raise ValueError("Please enter a positive number")
            return value
        except ValueError as e:
            print(f"Invalid input: {e}")
        except KeyboardInterrupt:
            print("\nOperation cancelled by user")
            return None

# Note: This won't run in notebook without actual input
# Uncomment to test with actual input:
# age = get_positive_integer("Enter your age: ")
# print(f"You entered: {age}")
````

### Database Operations
Handling database operations with exception handling.

````python
# Simulated database operation with exception handling
class DatabaseError(Exception):
    """Custom database error"""
    pass

def execute_query(query):
    try:
        if "DROP" in query.upper():
            raise DatabaseError("Dangerous operation detected")
        if not query:
            raise ValueError("Query cannot be empty")
        print(f"Executing: {query}")
        return "Query executed successfully"
    except DatabaseError as e:
        print(f"Database error: {e}")
        return None
    except ValueError as e:
        print(f"Validation error: {e}")
        return None

print("Test 1: Valid query")
execute_query("SELECT * FROM users")

print("\nTest 2: Dangerous query")
execute_query("DROP TABLE users")

print("\nTest 3: Empty query")
execute_query("")
````

## 8. Key Points & Summary

### Exception Hierarchy
- Base class: `BaseException`
- Main exception class: `Exception`
- All built-in exceptions inherit from `Exception`
- Custom exceptions should inherit from `Exception`

### Best Practices
✅ Catch specific exceptions
✅ Use meaningful error messages
✅ Use finally for cleanup operations
✅ Use context managers (with statement)
✅ Log exceptions for debugging
✅ Create custom exceptions for specific error conditions

### Common Pitfalls
❌ Catching generic `Exception` without reason
❌ Swallowing exceptions silently (empty except blocks)
❌ Using exceptions for normal flow control
❌ Not logging exception information
❌ Forgetting to clean up resources

### Key Takeaways
- Exceptions are Python's way of handling errors gracefully
- Use try-except blocks to handle expected exceptions
- Use else for code that runs only when no exception occurs
- Use finally for cleanup operations that must always run
- Raise exceptions for invalid conditions or inputs
- Create custom exceptions for specific error conditions
- Follow best practices for robust and maintainable code
