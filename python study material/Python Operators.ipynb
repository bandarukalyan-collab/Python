# Python Operators

## 1. Introduction to Python Operators

### What are Operators?
- Operators are special symbols in Python that perform operations on variables and values
- They are the building blocks of expressions and computations
- Python supports various types of operators for different operations

### Types of Operators:
1. Arithmetic Operators
2. Comparison (Relational) Operators
3. Logical Operators
4. Assignment Operators
5. Bitwise Operators
6. Identity Operators
7. Membership Operators

## 2. Arithmetic Operators

Arithmetic operators are used to perform mathematical operations.

| Operator | Name | Example |
|----------|------|---------|
| `+` | Addition | `10 + 5` |
| `-` | Subtraction | `10 - 5` |
| `*` | Multiplication | `10 * 5` |
| `/` | Division | `10 / 5` |
| `//` | Floor Division | `10 // 3` |
| `%` | Modulus | `10 % 3` |
| `**` | Exponentiation | `2 ** 3` |

```python
# Arithmetic Operators Examples
a = 10
b = 3

print(f"Values: a = {a}, b = {b}")
print(f"Addition (+):         a + b = {a + b}")
print(f"Subtraction (-):      a - b = {a - b}")
print(f"Multiplication (*):   a * b = {a * b}")
print(f"Division (/):         a / b = {a / b}  (always returns float)")
print(f"Floor Division (//): a // b = {a // b}  (integer division)")
print(f"Modulus (%):          a % b = {a % b}  (remainder)")
print(f"Exponentiation (**): a ** b = {a ** b}  (power)")
```

```python
# Practical Example: Calculate Average
scores = [85, 90, 78, 92, 88]
average = sum(scores) / len(scores)
print(f"Scores: {scores}")
print(f"Average score: {average}")
```

## 3. Comparison (Relational) Operators

Comparison operators are used to compare two values.

| Operator | Name | Example |
|----------|------|---------|
| `==` | Equal to | `5 == 5` |
| `!=` | Not equal to | `5 != 3` |
| `>` | Greater than | `5 > 3` |
| `<` | Less than | `3 < 5` |
| `>=` | Greater than or equal | `5 >= 5` |
| `<=` | Less than or equal | `3 <= 5` |

**Key Points:**
- Always return `True` or `False`
- Can compare different data types (with caution)
- Chain comparisons: `1 < x < 10`

```python
# Comparison Operators Examples
x = 5
y = 10

print(f"Values: x = {x}, y = {y}")
print(f"Equal (==):           x == y is {x == y}")
print(f"Not Equal (!=):       x != y is {x != y}")
print(f"Greater Than (>):     x > y is {x > y}")
print(f"Less Than (<):        x < y is {x < y}")
print(f"Greater or Equal (>=): x >= x is {x >= x}")
print(f"Less or Equal (<=):   y <= y is {y <= y}")
```

```python
# Chained Comparison
z = 7
print(f"Chained Comparison: 5 < {z} < 10 is {5 < z < 10}")

# Practical Example: Age Validation
age = 25
print(f"\nPractical Example - Age {age}:")
print(f"  Is adult (>= 18): {age >= 18}")
print(f"  Is senior (>= 65): {age >= 65}")
print(f"  Is teen (13-19): {13 <= age <= 19}")
```

## 4. Logical Operators

Logical operators are used to combine conditional statements.

| Operator | Description | Example |
|----------|-------------|---------|
| `and` | Returns True if both operands are True | `True and False` |
| `or` | Returns True if at least one operand is True | `True or False` |
| `not` | Reverses the boolean value | `not True` |

**Truth Table for AND:**
- `True and True = True`
- `True and False = False`
- `False and True = False`
- `False and False = False`

**Truth Table for OR:**
- `True or True = True`
- `True or False = True`
- `False or True = True`
- `False or False = False`

```python
# Logical Operators Examples
p = True
q = False

print(f"Values: p = {p}, q = {q}")
print(f"AND (p and q):       {p and q}  (True only if both are True)")
print(f"OR (p or q):         {p or q}  (True if at least one is True)")
print(f"NOT (not p):         {not p}  (Reverses the value)")
print(f"NOT (not q):         {not q}")
```

```python
# Truth Table Demonstration
print("Truth Table for AND:")
print(f"  True and True   = {True and True}")
print(f"  True and False  = {True and False}")
print(f"  False and True  = {False and True}")
print(f"  False and False = {False and False}")

print("\nTruth Table for OR:")
print(f"  True or True    = {True or True}")
print(f"  True or False   = {True or False}")
print(f"  False or True   = {False or True}")
print(f"  False or False  = {False or False}")
```

```python
# Practical Example: User Authentication
username = "admin"
password = "secret"
is_admin = username == "admin"
has_password = password == "secret"

print(f"Practical Example - Authentication:")
print(f"  Username valid: {is_admin}")
print(f"  Password valid: {has_password}")
print(f"  Access granted: {is_admin and has_password}")
```

## 5. Assignment Operators

Assignment operators are used to assign values to variables.

| Operator | Example | Equivalent To |
|----------|---------|---------------|
| `=` | `x = 5` | `x = 5` |
| `+=` | `x += 5` | `x = x + 5` |
| `-=` | `x -= 5` | `x = x - 5` |
| `*=` | `x *= 5` | `x = x * 5` |
| `/=` | `x /= 5` | `x = x / 5` |
| `//=` | `x //= 5` | `x = x // 5` |
| `%=` | `x %= 5` | `x = x % 5` |
| `**=` | `x **= 5` | `x = x ** 5` |

```python
# Assignment Operators Examples
num = 10
print(f"Initial value: num = {num}")

num += 5
print(f"After num += 5:  num = {num}  (equivalent to num = num + 5)")

num -= 3
print(f"After num -= 3:  num = {num}  (equivalent to num = num - 3)")

num *= 2
print(f"After num *= 2:  num = {num}  (equivalent to num = num * 2)")

num /= 4
print(f"After num /= 4:  num = {num}  (equivalent to num = num / 4)")
```

```python
# More Assignment Operators
num = 10
num //= 3
print(f"Reset to 10, then num //= 3:  num = {num}")

num = 10
num %= 3
print(f"Reset to 10, then num %= 3:   num = {num}")

num = 2
num **= 3
print(f"Reset to 2, then num **= 3:   num = {num}  (2^3 = 8)")
```

```python
# Practical Example: Counter
count = 0
print("Practical Example - Counter:")
for i in range(5):
    count += 1
print(f"  Count after 5 iterations: {count}")
```

## 6. Bitwise Operators

Bitwise operators work on binary representations of numbers.

| Operator | Name | Description | Example |
|----------|------|-------------|---------|
| `&` | AND | Sets bit to 1 if both bits are 1 | `5 & 3 = 1` |
| `|` | OR | Sets bit to 1 if at least one bit is 1 | `5 | 3 = 7` |
| `^` | XOR | Sets bit to 1 if bits are different | `5 ^ 3 = 6` |
| `~` | NOT | Inverts all bits | `~5 = -6` |
| `<<` | Left Shift | Shifts bits left | `5 << 1 = 10` |
| `>>` | Right Shift | Shifts bits right | `5 >> 1 = 2` |

**Use Cases:**
- Low-level programming
- Encryption algorithms
- Flag manipulation
- Performance optimization

```python
# Bitwise Operators Examples
m = 5  # Binary: 0101
n = 3  # Binary: 0011

print(f"Values: m = {m} (binary: {bin(m)}), n = {n} (binary: {bin(n)})")
print(f"Bitwise AND (&):     m & n = {m & n}  (binary: {bin(m & n)})")
print(f"Bitwise OR (|):      m | n = {m | n}  (binary: {bin(m | n)})")
print(f"Bitwise XOR (^):     m ^ n = {m ^ n}  (binary: {bin(m ^ n)})")
print(f"Bitwise NOT (~):     ~m = {~m}  (inverts all bits)")
print(f"Left Shift (<<):     m << 1 = {m << 1}  (binary: {bin(m << 1)})")
print(f"Right Shift (>>):    m >> 1 = {m >> 1}  (binary: {bin(m >> 1)})")
```

## 7. Identity Operators

Identity operators are used to check if two variables point to the same object.

| Operator | Description | Example |
|----------|-------------|---------|
| `is` | Returns True if both variables point to same object | `x is y` |
| `is not` | Returns True if variables point to different objects | `x is not y` |

**Key Points:**
- Check memory location (identity), not value
- Different from `==` operator
- Important for `None`, singleton objects

```python
# Identity Operators Examples
list1 = [1, 2, 3]
list2 = list1
list3 = [1, 2, 3]

print(f"list1 = {list1}")
print(f"list2 = list1 (same object)")
print(f"list3 = [1, 2, 3] (different object, same values)")

print(f"\nlist1 is list2:     {list1 is list2}  (same object in memory)")
print(f"list1 is list3:     {list1 is list3}  (different objects)")
print(f"list1 == list3:     {list1 == list3}  (same values)")
print(f"list1 is not list3: {list1 is not list3}")
```

```python
# Practical Example: Checking for None
result = None
print(f"Practical Example - None check:")
print(f"  result is None: {result is None}")
print(f"  result is not None: {result is not None}")

# String interning (small strings are cached)
str1 = "hello"
str2 = "hello"
print(f"\nString interning (small strings):")
print(f"  str1 = 'hello', str2 = 'hello'")
print(f"  str1 is str2: {str1 is str2}  (same object due to interning)")
```

## 8. Membership Operators

Membership operators are used to test if a sequence is present in an object.

| Operator | Description | Example |
|----------|-------------|---------|
| `in` | Returns True if value is found in sequence | `5 in [1, 2, 5]` |
| `not in` | Returns True if value is not found in sequence | `5 not in [1, 2, 3]` |

**Works With:**
- Lists, Tuples, Strings, Sets
- Dictionaries (checks keys)

```python
# Membership Operators Examples
fruits = ['apple', 'banana', 'cherry']
text = "Python programming"
person = {'name': 'John', 'age': 30}

print(f"fruits = {fruits}")
print(f"text = '{text}'")
print(f"person = {person}")

print(f"\n'apple' in fruits:     {'apple' in fruits}")
print(f"'grape' in fruits:     {'grape' in fruits}")
print(f"'grape' not in fruits: {'grape' not in fruits}")

print(f"\n'Python' in text:     {'Python' in text}")
print(f"'python' in text:     {'python' in text}  (case-sensitive)")
print(f"'xyz' not in text:    {'xyz' not in text}")
```

```python
# Membership with Dictionaries
print(f"'name' in person:     {'name' in person}  (checks keys)")
print(f"'John' in person:     {'John' in person}  (not a key)")

# Practical Example: Input Validation
allowed_colors = ['red', 'green', 'blue', 'yellow']
user_color = 'green'
print(f"\nPractical Example - Input validation:")
print(f"  Allowed colors: {allowed_colors}")
print(f"  User chose: '{user_color}'")
print(f"  Is valid: {user_color in allowed_colors}")
```

## 9. Operator Precedence

Operator precedence determines the order in which operators are evaluated.

**Order of Evaluation (Highest to Lowest):**
1. Parentheses `()`
2. Exponentiation `**`
3. Unary operators `+x`, `-x`, `~x`
4. Multiplication, Division, Floor Division, Modulus `*`, `/`, `//`, `%`
5. Addition, Subtraction `+`, `-`
6. Bitwise Shift `<<`, `>>`
7. Bitwise AND `&`
8. Bitwise XOR `^`
9. Bitwise OR `|`
10. Comparison operators `==`, `!=`, `>`, `<`, `>=`, `<=`
11. Identity operators `is`, `is not`
12. Membership operators `in`, `not in`
13. Logical NOT `not`
14. Logical AND `and`
15. Logical OR `or`

**Tip:** Use parentheses to make code clear and override precedence when needed.

```python
# Operator Precedence Examples
print("Example 1: Without parentheses")
result1 = 10 + 5 * 2
print(f"10 + 5 * 2 = {result1}  (multiplication first)")

print("\nExample 2: With parentheses")
result2 = (10 + 5) * 2
print(f"(10 + 5) * 2 = {result2}  (addition first)")

print("\nExample 3: Complex expression")
result3 = 10 + 5 * 2 ** 3 // 4
print(f"10 + 5 * 2 ** 3 // 4 = {result3}")
print("  Order: ** (exponent) -> * (multiply) -> // (floor divide) -> + (add)")
```

```python
# Example 4: Logical operators with comparison
result4 = 5 > 3 and 10 < 20
print(f"\nExample 4: Logical operators with comparison")
print(f"5 > 3 and 10 < 20 = {result4}")
```

## 10. Practical Real-World Examples

### Example 1: BMI Calculator
Calculate Body Mass Index using arithmetic operators.

```python
# BMI Calculator
weight = 70  # kg
height = 1.75  # meters
bmi = weight / (height ** 2)

print(f"Weight: {weight} kg, Height: {height} m")
print(f"BMI: {bmi:.2f}")

if bmi < 18.5:
    status = "Underweight"
elif 18.5 <= bmi < 25:
    status = "Normal"
elif 25 <= bmi < 30:
    status = "Overweight"
else:
    status = "Obese"
print(f"Status: {status}")
```

### Example 2: Shopping Cart Total
Calculate total price with discount and tax using arithmetic operators.

```python
# Shopping Cart
prices = [10.99, 5.50, 3.25, 15.00]
discount = 0.10  # 10% discount
tax_rate = 0.08  # 8% tax

subtotal = sum(prices)
discount_amount = subtotal * discount
after_discount = subtotal - discount_amount
tax_amount = after_discount * tax_rate
total = after_discount + tax_amount

print(f"Items: {prices}")
print(f"Subtotal: ${subtotal:.2f}")
print(f"Discount (10%): -${discount_amount:.2f}")
print(f"After discount: ${after_discount:.2f}")
print(f"Tax (8%): +${tax_amount:.2f}")
print(f"Total: ${total:.2f}")
```

### Example 3: Password Strength Checker
Check password strength using logical and membership operators.

```python
# Password Strength Checker
password = "Secure123!"
has_upper = any(c.isupper() for c in password)
has_lower = any(c.islower() for c in password)
has_digit = any(c.isdigit() for c in password)
has_special = any(c in "!@#$%^&*" for c in password)
is_long = len(password) >= 8

if has_upper and has_lower and has_digit and has_special and is_long:
    strength = "Strong"
elif (has_upper or has_lower) and has_digit and is_long:
    strength = "Medium"
else:
    strength = "Weak"

print(f"Password: '{password}'")
print(f"Has uppercase: {has_upper}")
print(f"Has lowercase: {has_lower}")
print(f"Has digit: {has_digit}")
print(f"Has special char: {has_special}")
print(f"Is 8+ chars: {is_long}")
print(f"Strength: {strength}")
```
