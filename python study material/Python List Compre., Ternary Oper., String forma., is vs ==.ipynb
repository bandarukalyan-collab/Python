# String Formatting

Python provides several ways to format strings. Each method has its own use cases and advantages.

### 1. f-strings (Formatted String Literals) - Python 3.6+

The most modern and recommended way to format strings.

````python
# Basic f-string
name = "Alice"
age = 25
print(f"My name is {name} and I'm {age} years old.")
````

````python
# f-string with expressions
a = 10
b = 20
print(f"The sum of {a} and {b} is {a + b}")
````

````python
# f-string with formatting options
pi = 3.14159
print(f"Pi rounded to 2 decimal places: {pi:.2f}")
print(f"Pi in scientific notation: {pi:.2e}")
````

````python
# f-string with padding and alignment
name = "Alice"
print(f"|{name:<10}|")  # Left-aligned
print(f"|{name:^10}|")  # Center-aligned
print(f"|{name:>10}|")  # Right-aligned
````

### 2. format() Method

Available in Python 2.7+ and Python 3.x

````python
# Basic format() method
name = "Bob"
age = 30
print("My name is {} and I'm {} years old.".format(name, age))
````

````python
# format() with positional arguments
print("{0} is {1} years old. {0} is a student.".format("Charlie", 22))
````

````python
# format() with keyword arguments
print("{name} is {age} years old.".format(name="David", age=28))
````

````python
# format() with formatting options
price = 1234.56789
print("Price: {:.2f}".format(price))
print("Price: {:,.2f}".format(price))  # With thousands separator
````

### 3. % Operator (Old Style)

The original string formatting method, still supported for backward compatibility.

````python
# Basic % formatting
name = "Eve"
age = 35
print("My name is %s and I'm %d years old." % (name, age))
````

````python
# % formatting with dictionary
data = {"name": "Frank", "age": 40}
print("%(name)s is %(age)d years old." % data)
````

````python
# % formatting with float precision
pi = 3.14159
print("Pi: %.2f" % pi)
````

# Identity vs Equality

Understanding the difference between `is` (identity) and `==` (equality) is crucial in Python.

### Equality Operator (==)

Checks if two objects have the **same value**.

````python
# Equality example
a = [1, 2, 3]
b = [1, 2, 3]

print(f"a == b: {a == b}")  # True - same values
print(f"a is b: {a is b}")    # False - different objects
````

### Identity Operator (is)

Checks if two variables refer to the **same object** in memory.

````python
# Identity example
a = [1, 2, 3]
b = a  # b references the same object as a

print(f"a == b: {a == b}")  # True - same values
print(f"a is b: {a is b}")    # True - same object

b[0] = 99
print(f"a: {a}")  # a is also changed because b and a reference the same object
````

### Small Integer Caching

Python caches small integers (-5 to 256) for performance.

````python
# Small integer caching
a = 100
b = 100
print(f"a == b: {a == b}")  # True
print(f"a is b: {a is b}")    # True (cached)

c = 300
d = 300
print(f"c == d: {c == d}")  # True
print(f"c is d: {c is d}")    # False (not cached)
````

### String Interning

Python may intern (reuse) strings that are literals or look like identifiers.

````python
# String interning
a = "hello"
b = "hello"
print(f"a == b: {a == b}")  # True
print(f"a is b: {a is b}")    # True (interned)

c = "hello world!"
d = "hello world!"
print(f"c == d: {c == d}")  # True
print(f"c is d: {c is d}")    # False (may not be interned)
````

### None Comparison

Always use `is` when comparing with `None`.

````python
# Comparing with None
x = None

# Correct way
if x is None:
    print("x is None")

# Also works but not recommended
if x == None:
    print("x equals None")
````

## Key Points: String Formatting & Identity vs Equality

### String Formatting:
- **f-strings**: Modern, readable, and recommended (Python 3.6+)
- **format()**: Flexible, works in older Python versions
- **% operator**: Legacy method, use only for backward compatibility

### Identity vs Equality:
- **==** checks if values are equal
- **is** checks if objects are the same in memory
- Use `is` only when you need to check object identity (e.g., with `None`)
- Use `==` for value comparisons in most cases
- Be aware of Python's caching behavior for small integers and string literals

# Ternary Operator

A ternary operator allows you to write conditional expressions in a single line.

## Syntax
```python
value_if_true if condition else value_if_false
```

````python
# Basic Example
age = 20
status = "adult" if age >= 18 else "minor"
print(f"Age: {age}, Status: {status}")  # adult
````

````python
# Example with numbers
x = 10
y = 20
max_value = x if x > y else y
print(f"Max of {x} and {y}: {max_value}")
````

````python
# Example with function calls
def greet(name):
    return f"Hello, {name}!"

message = greet("Alice") if True else greet("Bob")
print(message)
````

````python
# Nested ternary operator
score = 85
grade = "A" if score >= 90 else "B" if score >= 80 else "C"
print(f"Score: {score}, Grade: {grade}")
````

````python
# With None handling
name = None
display_name = name if name is not None else "Anonymous"
print(f"Display name: {display_name}")
````

````python
# In list comprehension
numbers = [1, 2, 3, 4, 5]
even_odd = ["even" if n % 2 == 0 else "odd" for n in numbers]
print(f"Numbers: {numbers}")
print(f"Even/Odd: {even_odd}")
````

````python
# In lambda function
get_status = lambda x: "positive" if x > 0 else "negative" if x < 0 else "zero"
print(f"Status of 5: {get_status(5)}")
print(f"Status of -3: {get_status(-3)}")
print(f"Status of 0: {get_status(0)}")
````

````python
# Practical example: Setting default values
user_input = ""
value = user_input if user_input else "default value"
print(f"Value: {value}")
````

# List Comprehension Examples

List comprehension provides a concise way to create lists based on existing lists or iterables.

````python
# Basic List Comprehension
# Traditional way
squares_traditional = []
for x in range(10):
    squares_traditional.append(x**2)
print(f"Traditional: {squares_traditional}")

# Using list comprehension
squares = [x**2 for x in range(10)]
print(f"List comprehension: {squares}")
````

````python
# List Comprehension with Condition (Filtering)
# Get even numbers from 0-9
evens = [x for x in range(10) if x % 2 == 0]
print(f"Even numbers: {evens}")

# Get numbers greater than 5
greater_than_five = [x for x in range(10) if x > 5]
print(f"Numbers > 5: {greater_than_five}")
````

````python
# List Comprehension with String Manipulation
words = ["hello", "world", "python", "programming"]
uppercase_words = [word.upper() for word in words]
print(f"Original: {words}")
print(f"Uppercase: {uppercase_words}")

# Get lengths of words
word_lengths = [len(word) for word in words]
print(f"Word lengths: {word_lengths}")
````

````python
# Nested List Comprehension (2D Matrix)
# Create a 3x3 multiplication table
matrix = [[i*j for j in range(1, 4)] for i in range(1, 4)]
print("Multiplication table:")
for row in matrix:
    print(row)
````

````python
# Combining Ternary Operator with List Comprehension
# Classify numbers as positive, negative, or zero
numbers = [-5, -3, 0, 2, 7, 10]
classifications = ["positive" if x > 0 else "negative" if x < 0 else "zero" for x in numbers]
print(f"Numbers: {numbers}")
print(f"Classifications: {classifications}")
````

````python
# Dictionary Comprehension
# Create a dictionary mapping numbers to their squares
squares_dict = {x: x**2 for x in range(6)}
print(f"Squares dictionary: {squares_dict}")

# Filter dictionary by values
filtered_dict = {k: v for k, v in squares_dict.items() if v > 10}
print(f"Filtered (values > 10): {filtered_dict}")
````

````python
# Set Comprehension
# Create a set of unique squares
squares_set = {x**2 for x in range(6)}
print(f"Squares set: {squares_set}")

# Get unique vowels from a string
text = "hello world"
vowels = {char for char in text if char in "aeiou"}
print(f"Vowels in '{text}': {vowels}")
````

````python
# Performance Comparison: List Comprehension vs For Loop
import time

# Test with list comprehension
start = time.time()
squares_lc = [x**2 for x in range(1000000)]
lc_time = time.time() - start

# Test with traditional for loop
start = time.time()
squares_loop = []
for x in range(1000000):
    squares_loop.append(x**2)
loop_time = time.time() - start

print(f"List comprehension time: {lc_time:.4f}s")
print(f"For loop time: {loop_time:.4f}s")
print(f"List comprehension is {loop_time/lc_time:.2f}x faster")
````

## Key Points: Ternary Operator

- The `if` condition comes **after** the true value (unlike other languages)
- Always requires both `if` and `else` parts
- Can be nested, but readability suffers with complex nesting
- Use parentheses for clarity in complex expressions
- Prefer traditional if-else for complex logic
