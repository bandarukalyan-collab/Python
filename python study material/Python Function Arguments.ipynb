# Python Function Arguments

## 1. Introduction to Function Arguments

### What are Function Arguments?
- Arguments are values passed to functions when calling them
- They allow functions to be flexible and reusable with different inputs
- Python supports multiple types of argument passing mechanisms

### Types of Function Arguments:
1. Default Arguments
2. Positional Arguments
3. Keyword Arguments
4. *args (Variable Positional Arguments)
5. **kwargs (Variable Keyword Arguments)

## 2. Default Arguments

Default arguments have predefined values that are used if no argument is passed.

**Syntax:**
```python
def function_name(param=default_value):
    # code
```

**Key Points:**
- Default values are assigned in the function definition
- If caller doesn't provide a value, the default is used
- Non-default parameters must come before default parameters
- Avoid mutable default values (like lists, dicts) - use None instead

````python
# Default Arguments Examples
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

print("Example 1: Using default greeting")
print(greet("Alice"))  # Uses default: "Hello, Alice!"

print("\nExample 2: Overriding default greeting")
print(greet("Bob", "Hi"))  # Overrides: "Hi, Bob!"

print("\nExample 3: Multiple default arguments")
def describe_person(name, age="unknown", city="unknown"):
    return f"{name} is {age} years old from {city}"

print(describe_person("Charlie"))
print(describe_person("David", "30"))
print(describe_person("Emma", "25", "London"))
````

````python
# Example: Default argument with mutable object (WARNING!)
def add_item_bad(item, cart=[]):
    cart.append(item)
    return cart

print("BAD - Mutable default argument:")
print(add_item_bad("apple"))  # ['apple']
print(add_item_bad("banana"))  # ['apple', 'banana'] - unexpected!

# CORRECT way: Use None as default
def add_item_good(item, cart=None):
    if cart is None:
        cart = []
    cart.append(item)
    return cart

print("\nGOOD - None as default:")
print(add_item_good("apple"))  # ['apple']
print(add_item_good("banana"))  # ['banana'] - correct!
````

## 3. Positional and Keyword Arguments

### Positional Arguments
- Arguments are matched based on their position in the function call
- Order matters - first argument goes to first parameter, etc.

### Keyword Arguments
- Arguments are passed using the parameter name
- Order doesn't matter
- Makes code more readable

**Best Practice:** Use keyword arguments for optional parameters to improve readability.

````python
# Positional vs Keyword Arguments
def student_info(name, age, grade):
    return f"Student: {name}, Age: {age}, Grade: {grade}"

print("Positional arguments (order matters):")
print(student_info("John", 15, "10th"))

print("\nKeyword arguments (order doesn't matter):")
print(student_info(name="Sarah", grade="11th", age=16))
print(student_info(age=17, grade="12th", name="Mike"))

print("\nMixed (positional first, then keyword):")
print(student_info("Emma", 14, grade="9th"))
````

````python
# Practical Example: Function with many parameters
def send_email(to, subject, body, cc=None, bcc=None, priority="normal"):
    print(f"To: {to}")
    print(f"Subject: {subject}")
    print(f"Body: {body}")
    if cc:
        print(f"CC: {cc}")
    if bcc:
        print(f"BCC: {bcc}")
    print(f"Priority: {priority}")

print("Example 1: Required parameters only")
send_email("user@example.com", "Welcome", "Hello!")

print("\n" + "="*40 + "\n")
print("Example 2: Using keyword arguments for clarity")
send_email(
    to="admin@company.com",
    subject="Alert",
    body="System error detected",
    priority="high"
)
````

## 4. *args - Variable Positional Arguments

`*args` allows a function to accept any number of positional arguments.

**Syntax:**
```python
def function_name(*args):
    # args is a tuple
```

**Key Points:**
- `*args` collects extra positional arguments into a tuple
- The name `args` is a convention - you can use any name with `*`
- Must come after regular positional parameters
- Useful when you don't know how many arguments will be passed

````python
# *args Examples
def sum_all(*numbers):
    print(f"Received {len(numbers)} arguments: {numbers}")
    print(f"Type of args: {type(numbers)}")
    return sum(numbers)

print("Example 1: Summing multiple numbers")
result1 = sum_all(1, 2, 3, 4, 5)
print(f"Sum: {result1}\n")

print("Example 2: No arguments")
result2 = sum_all()
print(f"Sum: {result2}\n")

print("Example 3: Different number of arguments")
print(sum_all(10, 20))
print(sum_all(100, 200, 300, 400))
````

````python
# Combining regular parameters with *args
def greet_all(greeting, *names):
    for name in names:
        print(f"{greeting}, {name}!")

print("Example: Greeting multiple people")
greet_all("Hello", "Alice", "Bob", "Charlie")

print("\nExample: Custom greeting")
greet_all("Good morning", "David", "Emma")
````

## 5. **kwargs - Variable Keyword Arguments

`**kwargs` allows a function to accept any number of keyword arguments.

**Syntax:**
```python
def function_name(**kwargs):
    # kwargs is a dictionary
```

**Key Points:**
- `**kwargs` collects extra keyword arguments into a dictionary
- Keys are parameter names, values are argument values
- The name `kwargs` is a convention
- Must come after `*args` if both are used
- Useful for configuration options, flexible APIs

````python
# **kwargs Examples
def print_info(**info):
    print(f"Received {len(info)} keyword arguments:")
    print(f"Type of kwargs: {type(info)}")
    for key, value in info.items():
        print(f"  {key}: {value}")

print("Example 1: Personal information")
print_info(name="John", age=30, city="New York")

print("\nExample 2: No keyword arguments")
print_info()

print("\nExample 3: Configuration options")
print_info(debug=True, verbose=False, log_file="app.log")
````

````python
# Combining regular parameters with **kwargs
def create_user(username, **attributes):
    user = {"username": username}
    user.update(attributes)
    return user

print("Example: Creating user with attributes")
user1 = create_user("john_doe", email="john@example.com", age=25)
print(user1)

user2 = create_user("jane_smith", email="jane@example.com", 
                    age=30, city="London", active=True)
print(user2)
````

## 6. Combining *args and **kwargs

You can use both `*args` and `**kwargs` in the same function.

**Order of parameters (IMPORTANT):**
1. Regular positional parameters
2. Default parameters
3. `*args`
4. `**kwargs`

**Syntax:**
```python
def func(a, b=10, *args, **kwargs):
    pass
```

````python
# Combining all argument types
def complex_function(required, optional="default", *args, **kwargs):
    print(f"Required: {required}")
    print(f"Optional: {optional}")
    print(f"Args (*args): {args}")
    print(f"Kwargs (**kwargs): {kwargs}")

print("Example 1: All types")
complex_function("req_val", "opt_val", 1, 2, 3, x=10, y=20)

print("\n" + "="*40 + "\n")
print("Example 2: Only required and args")
complex_function("req_val", 1, 2, 3, 4)

print("\n" + "="*40 + "\n")
print("Example 3: Only required and kwargs")
complex_function("req_val", a=1, b=2)
````

````python
# Practical Example: Flexible function
def log_message(level, message, *extra_info, **options):
    timestamp = options.get("timestamp", "NOW")
    prefix = options.get("prefix", "LOG")
    
    print(f"[{timestamp}] [{prefix}] [{level}] {message}")
    
    if extra_info:
        print("Additional info:")
        for info in extra_info:
            print(f"  - {info}")

print("Example 1: Basic logging")
log_message("INFO", "Application started")

print("\n" + "="*40 + "\n")
print("Example 2: With extra info and options")
log_message("ERROR", "Database connection failed", 
           "Retry attempt 1", "Timeout: 30s",
           timestamp="2024-01-15 10:30:00", prefix="APP")
````

## 7. Unpacking Arguments

You can unpack sequences and dictionaries when calling functions.

### Unpacking Positional Arguments (*)
Use `*` to unpack a list/tuple into positional arguments.

### Unpacking Keyword Arguments (**)
Use `**` to unpack a dictionary into keyword arguments.

````python
# Unpacking positional arguments
def multiply(a, b, c):
    return a * b * c

numbers = [2, 3, 4]
print(f"List: {numbers}")
print(f"Unpacked result: {multiply(*numbers)}")  # multiply(2, 3, 4)

print("\n" + "="*40 + "\n")

# Unpacking keyword arguments
def describe_person(name, age, city):
    return f"{name}, {age}, {city}"

person = {"name": "Alice", "age": 30, "city": "NYC"}
print(f"Dictionary: {person}")
print(f"Unpacked result: {describe_person(**person)}")
````

````python
# Combining unpacking with regular arguments
def calculate(operation, a, b, c=0):
    if operation == "sum":
        return a + b + c
    elif operation == "product":
        return a * b * c
    return 0

values = [5, 10]
print("Example 1: Unpacking with regular argument")
print(calculate("sum", *values))  # calculate("sum", 5, 10)

print("\nExample 2: Unpacking with keyword argument")
print(calculate("sum", *values, c=15))  # calculate("sum", 5, 10, c=15)
````

## 8. Practical Real-World Examples

### Example 1: Flexible Calculator
A calculator that can handle any number of operations.

````python
# Flexible Calculator
def calculator(operation, *numbers):
    if not numbers:
        return "No numbers provided"
    
    if operation == "add":
        return sum(numbers)
    elif operation == "multiply":
        result = 1
        for num in numbers:
            result *= num
        return result
    elif operation == "max":
        return max(numbers)
    elif operation == "min":
        return min(numbers)
    else:
        return "Unknown operation"

print("Calculator Examples:")
print(f"Add: {calculator('add', 1, 2, 3, 4, 5)}")
print(f"Multiply: {calculator('multiply', 2, 3, 4)}")
print(f"Max: {calculator('max', 10, 5, 20, 15)}")
print(f"Min: {calculator('min', 10, 5, 20, 15)}")
````

### Example 2: Configuration Manager
A function that manages configuration with default and custom settings.

````python
# Configuration Manager
def get_config(**overrides):
    default_config = {
        "debug": False,
        "log_level": "INFO",
        "max_connections": 10,
        "timeout": 30,
        "cache_enabled": True
    }
    
    # Merge overrides with defaults
    config = default_config.copy()
    config.update(overrides)
    
    return config

print("Default configuration:")
print(get_config())

print("\nCustom configuration:")
print(get_config(debug=True, log_level="DEBUG", max_connections=50))
````

### Example 3: HTML Builder
Build HTML elements with flexible attributes.

````python
# HTML Builder
def build_html(tag, content="", **attributes):
    # Build attribute string
    attrs = ""
    for key, value in attributes.items():
        # Replace underscores with hyphens for HTML attributes
        attr_name = key.replace("_", "-")
        attrs += f' {attr_name}="{value}"'
    
    # Build HTML element
    if content:
        return f"<{tag}{attrs}>{content}</{tag}>"
    else:
        return f"<{tag}{attrs}></{tag}>"

print("HTML Examples:")
print(build_html("p", "Hello World"))
print(build_html("a", "Click me", href="https://example.com", target="_blank"))
print(build_html("input", type="text", placeholder="Enter name", id="name_input"))
print(build_html("div", "Content", class_name="container", id="main"))
````

## 9. Best Practices

### Do:
- Use keyword arguments for optional parameters (improves readability)
- Use `None` as default for mutable objects instead of empty lists/dicts
- Use `*args` when the number of arguments is truly variable
- Use `**kwargs` for flexible configuration options
- Document your function parameters clearly

### Don't:
- Don't use mutable default values directly
- Don't mix positional and keyword arguments confusingly
- Don't overuse `*args` and `**kwargs` when regular parameters would be clearer
- Don't forget the parameter order: regular, default, *args, **kwargs

````python
# Good vs Bad Examples

# BAD: Mutable default argument
def bad_append(item, lst=[]):
    lst.append(item)
    return lst

# GOOD: None as default
def good_append(item, lst=None):
    if lst is None:
        lst = []
    lst.append(item)
    return lst

# BAD: Unclear positional arguments
def bad_send_email(to, subject, body, cc, bcc, priority, format):
    pass

# GOOD: Keyword arguments for clarity
def good_send_email(to, subject, body, cc=None, bcc=None, 
                   priority="normal", format="html"):
    pass
````

## 10. Summary

| Argument Type | Syntax | Purpose | Example |
|--------------|--------|---------|---------|
| Default | `param=value` | Provide fallback values | `def greet(name, msg="Hi")` |
| Positional | `func(arg1, arg2)` | Match by order | `greet("Alice", "Hello")` |
| Keyword | `func(name=value)` | Match by name | `greet(name="Alice", msg="Hi")` |
| *args | `def func(*args)` | Variable positional | `def sum_all(*nums)` |
| **kwargs | `def func(**kwargs)` | Variable keyword | `def config(**opts)` |

**Parameter Order:**
```
def func(required_params, default_params, *args, **kwargs):
    pass
```
