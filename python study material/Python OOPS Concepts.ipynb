# Python OOPS Concepts

## 1. Introduction to OOPS

### What is OOPS?
- **OOPS**: Object-Oriented Programming System
- A programming paradigm based on "objects" containing data and methods
- Organizes software design around data, or objects, rather than functions and logic
- Python is a multi-paradigm language that supports OOPS

### OOPS Principles in Python
- **Class**: Blueprint for creating objects
- **Object**: Instance of a class
- **Encapsulation**: Bundling data and methods
- **Inheritance**: Deriving new classes from existing ones
- **Polymorphism**: Same interface, different implementations
- **Abstraction**: Hiding complex implementation details

### Why Use OOPS?
- **Modularity**: Code is organized into manageable pieces
- **Reusability**: Code can be reused through inheritance
- **Maintainability**: Easier to update and maintain
- **Scalability**: Easier to scale applications
- **Real-world modeling**: Maps real-world entities to code

## 2. Classes and Objects

### Defining Classes
- Use `class` keyword to define a class
- Class names should use CamelCase convention
- Classes can contain attributes (variables) and methods (functions)

```python
# Defining a simple class
class Car:
    # Class attribute
    wheels = 4
    
    # Constructor method
    def __init__(self, brand, model):
        # Instance attributes
        self.brand = brand
        self.model = model
    
    # Instance method
    def describe(self):
        return f"{self.brand} {self.model}"
    
    def honk(self):
        return "Beep beep!"

# Creating objects (instances)
car1 = Car("Toyota", "Corolla")
car2 = Car("Honda", "Civic")

print(f"Car 1: {car1.describe()}")
print(f"Car 2: {car2.describe()}")
print(f"Wheels: {car1.wheels}")
print(f"Honk: {car1.honk()}")
```

### __init__ Method
- Constructor method called when object is created
- Initializes object's attributes
- First parameter is always `self` (reference to current object)

```python
class Person:
    def __init__(self, name, age):
        print("__init__ called - Creating Person object")
        self.name = name
        self.age = age
    
    def introduce(self):
        return f"Hi, I'm {self.name} and I'm {self.age} years old"

# __init__ is called automatically
person = Person("Alice", 25)
print(person.introduce())
```

### Class Attributes vs Instance Attributes
- **Class attributes**: Shared by all instances
- **Instance attributes**: Unique to each instance

```python
class Dog:
    # Class attribute
    species = "Canis familiaris"
    
    def __init__(self, name, breed):
        # Instance attributes
        self.name = name
        self.breed = breed

dog1 = Dog("Buddy", "Golden Retriever")
dog2 = Dog("Max", "Labrador")

print(f"Dog 1: {dog1.name}, {dog1.breed}, {dog1.species}")
print(f"Dog 2: {dog2.name}, {dog2.breed}, {dog2.species}")

# Class attribute is shared
print(f"Same species object: {dog1.species is dog2.species}")
```

## 3. Encapsulation

### What is Encapsulation?
- Bundling data and methods that operate on the data within a single unit (class)
- Restricting direct access to some of an object's components
- Hiding internal state and requiring interaction through methods

### Access Modifiers in Python
- **Public**: Accessible from anywhere (default)
- **Protected**: Should not be accessed outside class (convention: `_var`)
- **Private**: Cannot be accessed outside class (convention: `__var`)

```python
class BankAccount:
    def __init__(self, balance):
        # Private attribute
        self.__balance = balance
        # Protected attribute
        self._owner = "Unknown"
        # Public attribute
        self.account_number = "123456"
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return True
        return False
    
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            return True
        return False
    
    def get_balance(self):
        return self.__balance

account = BankAccount(1000)
print(f"Initial balance: {account.get_balance()}")
account.deposit(500)
print(f"After deposit: {account.get_balance()}")
account.withdraw(200)
print(f"After withdrawal: {account.get_balance()}")
```

### Name Mangling
- Python changes names of private attributes to prevent accidental access
- `__var` becomes `_ClassName__var`
- Still accessible but indicates it should not be used

```python
# Name mangling demonstration
class MyClass:
    def __init__(self):
        self.__private = "private"
        self._protected = "protected"
        self.public = "public"

obj = MyClass()
print(f"Public: {obj.public}")
print(f"Protected: {obj._protected}")
print(f"Private (name mangled): {obj._MyClass__private}")
```

### Property Decorators
- Use `@property` to create getter methods
- Use `@attribute.setter` to create setter methods
- Provides controlled access to private attributes

```python
class Temperature:
    def __init__(self, celsius):
        self.__celsius = celsius
    
    @property
    def celsius(self):
        return self.__celsius
    
    @celsius.setter
    def celsius(self, value):
        if value < -273.15:
            raise ValueError("Temperature below absolute zero")
        self.__celsius = value
    
    @property
    def fahrenheit(self):
        return (self.__celsius * 9/5) + 32

temp = Temperature(25)
print(f"Celsius: {temp.celsius}")
print(f"Fahrenheit: {temp.fahrenheit:.2f}")

temp.celsius = 30
print(f"New Celsius: {temp.celsius}")
```

## 4. Inheritance

### What is Inheritance?
- Mechanism where a new class derives attributes and methods from an existing class
- Promotes code reuse
- Creates hierarchical relationships between classes

### Types of Inheritance
- **Single Inheritance**: One child class inherits from one parent class
- **Multiple Inheritance**: Child class inherits from multiple parent classes
- **Multilevel Inheritance**: Chain of inheritance (grandparent -> parent -> child)

### Single Inheritance

```python
# Single inheritance
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return "Some sound"
    
    def eat(self):
        return f"{self.name} is eating"

class Dog(Animal):
    def speak(self):
        return "Woof!"

dog = Dog("Buddy")
print(f"Name: {dog.name}")
print(f"Sound: {dog.speak()}")
print(f"Action: {dog.eat()}")
```

### Multiple Inheritance

```python
# Multiple inheritance
class Flyer:
    def fly(self):
        return "Flying"

class Swimmer:
    def swim(self):
        return "Swimming"

class Duck(Flyer, Swimmer):
    def quack(self):
        return "Quack!"

duck = Duck()
print(f"Duck can {duck.fly()}")
print(f"Duck can {duck.swim()}")
print(f"Duck says: {duck.quack()}")
```

### Multilevel Inheritance

```python
# Multilevel inheritance
class Grandparent:
    def grandparent_method(self):
        return "Grandparent method"

class Parent(Grandparent):
    def parent_method(self):
        return "Parent method"

class Child(Parent):
    def child_method(self):
        return "Child method"

child = Child()
print(child.grandparent_method())
print(child.parent_method())
print(child.child_method())
```

### Method Resolution Order (MRO)
- Order in which Python searches for methods in multiple inheritance
- Follows C3 linearization algorithm
- Can be viewed using `ClassName.mro()` or `ClassName.__mro__`

```python
# MRO demonstration
class A:
    def method(self):
        return "A"

class B(A):
    def method(self):
        return "B"

class C(A):
    def method(self):
        return "C"

class D(B, C):
    pass

obj = D()
print(f"Method result: {obj.method()}")
print(f"MRO: {D.mro()}")
```

### super() Function
- Used to call methods from parent class
- Essential for cooperative multiple inheritance
- Avoids hardcoding parent class names

```python
# Using super()
class Vehicle:
    def __init__(self, brand):
        self.brand = brand
    
    def start(self):
        return f"{self.brand} vehicle starting"

class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model
    
    def start(self):
        parent_start = super().start()
        return f"{parent_start} - {self.model}"

car = Car("Toyota", "Corolla")
print(car.start())
```

## 5. Polymorphism

### What is Polymorphism?
- "Many forms" - same interface, different implementations
- Allows objects of different classes to be treated as objects of a common superclass
- Key principle: "Program to an interface, not an implementation"

### Method Overriding
- Subclass provides specific implementation of method already defined in superclass

```python
# Method overriding
class Shape:
    def area(self):
        return "Area calculation not defined"

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14159 * self.radius ** 2

shapes = [Rectangle(5, 3), Circle(2)]
for shape in shapes:
    print(f"Area: {shape.area():.2f}")
```

### Method Overloading (Simulated)
- Python doesn't support traditional method overloading
- Can simulate using default arguments or *args/**kwargs

```python
# Simulated method overloading
class Calculator:
    def add(self, a, b=None, c=None):
        if b is None and c is None:
            return a
        elif c is None:
            return a + b
        else:
            return a + b + c

calc = Calculator()
print(f"add(5): {calc.add(5)}")
print(f"add(5, 3): {calc.add(5, 3)}")
print(f"add(5, 3, 2): {calc.add(5, 3, 2)}")
```

### Duck Typing
- "If it walks like a duck and quacks like a duck, it's a duck"
- Focuses on behavior rather than type
- No need for explicit inheritance

```python
# Duck typing
class Duck:
    def quack(self):
        return "Quack!"

class Person:
    def quack(self):
        return "I'm quacking like a duck!"

def make_quack(obj):
    return obj.quack()

duck = Duck()
person = Person()

print(make_quack(duck))
print(make_quack(person))
```

### Operator Overloading
- Define how operators behave with custom objects
- Use special methods (dunder methods)

```python
# Operator overloading
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
    
    def __str__(self):
        return f"Vector({self.x}, {self.y})"
    
    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = v1 + v2
print(f"v1 + v2 = {v3}")
print(f"v1 == v2: {v1 == v2}")
```

## 6. Data Abstraction

### What is Abstraction?
- Hiding complex implementation details
- Showing only essential features
- Reducing complexity and increasing efficiency

### Abstract Classes
- Classes that cannot be instantiated
- Designed to be subclassed
- Can contain abstract methods (methods without implementation)

```python
from abc import ABC, abstractmethod

# Abstract class
class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass
    
    @abstractmethod
    def move(self):
        pass

# Concrete classes must implement abstract methods
class Dog(Animal):
    def make_sound(self):
        return "Woof!"
    
    def move(self):
        return "Running"

class Cat(Animal):
    def make_sound(self):
        return "Meow!"
    
    def move(self):
        return "Prowling"

# Cannot instantiate abstract class
# animal = Animal()  # This would raise TypeError

dog = Dog()
cat = Cat()

print(f"Dog: {dog.make_sound()}, {dog.move()}")
print(f"Cat: {cat.make_sound()}, {cat.move()}")
```

### Interface Implementation
- Python doesn't have explicit interfaces
- Use abstract classes with only abstract methods
- Protocol classes (Python 3.8+) for structural subtyping

```python
# Interface-like pattern
class ShapeInterface(ABC):
    @abstractmethod
    def area(self):
        pass
    
    @abstractmethod
    def perimeter(self):
        pass

class Rectangle(ShapeInterface):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)

rect = Rectangle(5, 3)
print(f"Area: {rect.area()}")
print(f"Perimeter: {rect.perimeter()}")
```

## 7. Meta Classes

### What are Meta Classes?
- "Class of a class" - classes that create classes
- Control how classes are created and behave
- Default metaclass in Python is `type`
- Advanced feature for metaprogramming

### type() Function
- Can be used to create classes dynamically
- `type(name, bases, dict)` syntax

```python
# Creating class using type()
def __init__(self, name):
    self.name = name

def greet(self):
    return f"Hello, {self.name}!"

# Dynamically create a class
Person = type('Person', (object), {
    '__init__': __init__,
    'greet': greet
})

person = Person("Alice")
print(person.greet())
print(f"Type of Person: {type(Person)}")  # <class 'type'>
```

### Creating Custom Meta Classes
- Inherit from `type`
- Override `__new__` or `__init__` methods
- Can modify class creation process

```python
# Custom metaclass
class SingletonMeta(type):
    _instances = {}
    
    def __call__(cls, *args, **kwargs):
        if cls not in cls._instances:
            instance = super().__call__(*args, **kwargs)
            cls._instances[cls] = instance
        return cls._instances[cls]

class Database(metaclass=SingletonMeta):
    def __init__(self):
        print("Database connection initialized")

# Both references point to same instance
db1 = Database()
db2 = Database()
print(f"Same instance: {db1 is db2}")
```

### When to Use Meta Classes
- **Singleton pattern**: Ensure only one instance exists
- **Class registration**: Keep track of all subclasses
- **Automatic validation**: Validate class attributes
- **API modification**: Change class behavior globally

### When NOT to Use Meta Classes
- Simple class modifications (use decorators instead)
- When you're not sure what you're doing
- When simpler solutions exist

## 8. Monkey Patching

### What is Monkey Patching?
- Dynamically modifying code at runtime
- Adding or replacing methods/functions in existing classes or modules
- Can be done to third-party libraries without modifying source code
- Named because it's "messing with" code

### Runtime Modification
- Python allows modifying classes and functions at runtime
- Can add, modify, or delete attributes and methods

```python
# Monkey patching example
class Dog:
    def bark(self):
        return "Woof!"

# Original behavior
dog = Dog()
print(f"Original: {dog.bark()}")

# Monkey patch - add new method
def meow(self):
    return "Meow! (I'm a confused dog)"

Dog.meow = meow
print(f"Patched: {dog.meow()}")

# Monkey patch - replace existing method
def new_bark(self):
    return "BARK BARK!"

Dog.bark = new_bark
print(f"Replaced: {dog.bark()}")
```

### Use Cases for Monkey Patching
- **Testing**: Replace external dependencies with mocks
- **Bug fixes**: Temporarily fix bugs in third-party libraries
- **Feature addition**: Add features to existing libraries
- **Performance optimization**: Replace slow implementations

```python
# Practical example: Patching for testing
import math

# Original sqrt function
print(f"Original sqrt(16): {math.sqrt(16)}")

# Monkey patch for testing (e.g., to simulate error)
original_sqrt = math.sqrt

def patched_sqrt(x):
    if x < 0:
        raise ValueError("Cannot calculate square root of negative number")
    return original_sqrt(x)

math.sqrt = patched_sqrt
print(f"Patched sqrt(16): {math.sqrt(16)}")
```

### Risks and Best Practices

**Risks:**
- Can make code hard to understand and debug
- Changes affect all instances globally
- Can break when libraries are updated
- Not thread-safe in most cases

**Best Practices:**
- Use sparingly and only when necessary
- Document patches clearly
- Restore original behavior when done
- Consider alternatives (decorators, inheritance)
- Use in testing environments only

```python
# Safe monkey patching with restoration
class StringOperations:
    def reverse(self, s):
        return s[::-1]

# Save original method
original_reverse = StringOperations.reverse

# Apply patch
def enhanced_reverse(self, s):
    result = original_reverse(self, s)
    return f"Reversed: {result}"

StringOperations.reverse = enhanced_reverse

# Test patched version
ops = StringOperations()
print(ops.reverse("hello"))

# Restore original
StringOperations.reverse = original_reverse
print(f"Restored: {ops.reverse('hello')}")
```

## 9. Practical Examples

### Real-world OOPS Scenario: Employee Management System

```python
# Employee management system
from abc import ABC, abstractmethod

class Employee(ABC):
    def __init__(self, name, employee_id):
        self.__name = name
        self.__employee_id = employee_id
    
    @property
    def name(self):
        return self.__name
    
    @property
    def employee_id(self):
        return self.__employee_id
    
    @abstractmethod
    def calculate_salary(self):
        pass

class FullTimeEmployee(Employee):
    def __init__(self, name, employee_id, monthly_salary):
        super().__init__(name, employee_id)
        self.__monthly_salary = monthly_salary
    
    def calculate_salary(self):
        return self.__monthly_salary

class PartTimeEmployee(Employee):
    def __init__(self, name, employee_id, hourly_rate, hours_worked):
        super().__init__(name, employee_id)
        self.__hourly_rate = hourly_rate
        self.__hours_worked = hours_worked
    
    def calculate_salary(self):
        return self.__hourly_rate * self.__hours_worked

# Polymorphism in action
employees = [
    FullTimeEmployee("Alice", "E001", 5000),
    FullTimeEmployee("Bob", "E002", 6000),
    PartTimeEmployee("Charlie", "E003", 20, 80),
    PartTimeEmployee("David", "E004", 25, 60)
]

for emp in employees:
    print(f"{emp.name} (ID: {emp.employee_id}): ${emp.calculate_salary():,.2f}")
```
