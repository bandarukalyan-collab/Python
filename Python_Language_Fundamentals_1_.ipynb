4. Data Types:

    Below are the inbuilt datatypes:

    int, float, complex, bool, str, bytes, bytearray,
    range, list, tuple, set, frozenset, dict, none

    type() --> to check the type of variable
    
    id() --> to get address of object

    Slice Operator:

    *   [] operator is called slice operator, which can be used to retrieve parts of string.
    *   +ve index means forward direction from left to right. -ve index means backward direction from right to left.

```python
#Datatypes Python#

my_int = 10
print(f"Value: {my_int}, Type: {type(my_int)}")

my_float = 10.5
print(f"Value: {my_float}, Type: {type(my_float)}")

my_complex = 3 + 4j
print(f"Value: {my_complex}, Type: {type(my_complex)}")

my_bool = True
print(f"Value: {my_bool}, Type: {type(my_bool)}")

my_str = "Hello, Python!"
print(f"Value: {my_str}, Type: {type(my_str)}")

my_bytes = b"hello"
print(f"Value: {my_bytes}, Type: {type(my_bytes)}")

my_bytearray = bytearray(b"world")
print(f"Value: {my_bytearray}, Type: {type(my_bytearray)}")

my_range = range(5)
print(f"Value: {my_range}, Type: {type(my_range)}")

my_list = [1, 2, 3]
print(f"Value: {my_list}, Type: {type(my_list)}")

my_tuple = (1, 2, 3)
print(f"Value: {my_tuple}, Type: {type(my_tuple)}")

my_set = {1, 2, 3}
print(f"Value: {my_set}, Type: {type(my_set)}")

my_frozenset = frozenset({1, 2, 3})
print(f"Value: {my_frozenset}, Type: {type(my_frozenset)}")

my_dict = {'a': 1, 'b': 2}
print(f"Value: {my_dict}, Type: {type(my_dict)}")

my_none = None
print(f"Value: {my_none}, Type: {type(my_none)}")

a = 10
b = 10
c = 20

print(f"\nValue of a: {a}, Type of a: {type(a)}, ID of a: {id(a)}")
print(f"Value of b: {b}, Type of b: {type(b)}, ID of b: {id(b)}")
print(f"Value of c: {c}, Type of c: {type(c)}, ID of c: {id(c)}")

s1 = "hello"
s2 = "hello"

print(f"\nValue of s1: {s1}, Type of s1: {type(s1)}, ID of s1: {id(s1)}")
print(f"Value of s2: {s2}, Type of s2: {type(s2)}, ID of s2: {id(s2)}")

# Slice Operator Eg:

s = "Python is Object Oriented Programming Language"
s[1:27]
```

  5. Type Casting:

        1.   int()
        2.   float()
        3.   complex()
        4.   bool
        5.   str()

```python
# int() typecasting
num_str = "123"
num_float = 10.5

int_from_str = int(num_str)
int_from_float = int(num_float)

print(f"int('123'): {int_from_str}, Type: {type(int_from_str)}")
print(f"int(10.5): {int_from_float}, Type: {type(int_from_float)}")

# float() typecasting
num_int = 25
num_str_float = "3.14"

float_from_int = float(num_int)
float_from_str = float(num_str_float)

print(f"float(25): {float_from_int}, Type: {type(float_from_int)}")
print(f"float('3.14'): {float_from_str}, Type: {type(float_from_str)}")

# complex() typecasting
num_real = 5
num_imag = 2.5
num_str_complex = "1+2j"

complex_from_int = complex(num_real)
complex_from_parts = complex(num_real, num_imag)
complex_from_str = complex(num_str_complex)

print(f"complex(5): {complex_from_int}, Type: {type(complex_from_int)}")
print(f"complex(5, 2.5): {complex_from_parts}, Type: {type(complex_from_parts)}")
print(f"complex('1+2j'): {complex_from_str}, Type: {type(complex_from_str)}")

# bool() typecasting
int_zero = 0
int_nonzero = 100
empty_str = ""
non_empty_str = "hello"
empty_list = []
non_empty_list = [1]

bool_from_zero = bool(int_zero)
bool_from_nonzero = bool(int_nonzero)
bool_from_empty_str = bool(empty_str)
bool_from_non_empty_str = bool(non_empty_str)
bool_from_empty_list = bool(empty_list)
bool_from_non_empty_list = bool(non_empty_list)

print(f"bool(0): {bool_from_zero}, Type: {type(bool_from_zero)}")
print(f"bool(100): {bool_from_nonzero}, Type: {type(bool_from_nonzero)}")
print(f"bool(''): {bool_from_empty_str}, Type: {type(bool_from_empty_str)}")
print(f"bool('hello'): {bool_from_non_empty_str}, Type: {type(bool_from_non_empty_str)}")
print(f"bool([]): {bool_from_empty_list}, Type: {type(bool_from_empty_list)}")
print(f"bool([1]): {bool_from_non_empty_list}, Type: {type(bool_from_non_empty_list)}")

# str() typecasting
my_int_val = 456
my_float_val = 78.9
my_list_val = [10, 20]

str_from_int = str(my_int_val)
str_from_float = str(my_float_val)
str_from_list = str(my_list_val)

print(f"str(456): {str_from_int}, Type: {type(str_from_int)}")
print(f"str(78.9): {str_from_float}, Type: {type(str_from_float)}")
print(f"str([10, 20]): {str_from_list}, Type: {type(str_from_list)}")
```

bytes vs bytearray datatype:

| Feature      | bytes         |bytearray       
| ------------ | ----------------- | ------------------
| Mutability   | Immutable         | Mutable
| Syntax       | bytes[x]          | bytearray[x]
| Modification | Not allowed       | Allowed
| Use case     | Fixed binary data | Editable binary data
byte range : 0 to 256
